# Demoncrat

```python
path_dem = '/Users/ninashao/Documents/GitHub/fall2017-project1-NinaShao1230/data/inauguralsDem'
lee_train_file_dem = glob.glob(os.path.join(path_dem, '*.txt'))
```

```python
def build_texts(fname):
    for fle in glob.glob(os.path.join(fname, '*.txt')):
        f = open(fle,"r")
        for line in f:
            yield gensim.utils.simple_preprocess(line,deacc=True, min_len=3)
```

```python
train_texts_dem = list(build_texts(path_dem))
len(train_texts_dem)
```

```python
bigram_dem = gensim.models.Phrases(train_texts_dem)
stops = set(stopwords.words('english'))
```

```python
def process_texts(texts):
    texts = [[word for word in line if word not in stops] for line in texts]
    texts = [bigram[line] for line in texts]
    return texts
```

```python
train_texts_dem = process_texts(train_texts_dem)
```

```python
dictionary_dem = Dictionary(train_texts_dem)
corpus_dem = [dictionary_dem.doc2bow(text) for text in train_texts_dem]
```

```python
hdpmodel_dem = HdpModel(corpus=corpus_dem, id2word=dictionary_dem)
```

```python
hdpmodel_dem.show_topics()
```

```python
hdptopics_dem = hdpmodel_dem.show_topics(formatted=False)
hdptopics_dem
```
