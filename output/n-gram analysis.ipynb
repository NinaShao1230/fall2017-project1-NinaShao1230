# n-gram

```python
import collections
import string 
from collections import Counter 
from nltk.tokenize import RegexpTokenizer 
from nltk.corpus import stopwords
from nltk.util import ngrams

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer 
```

```python
stemmer = PorterStemmer()

def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
        return stemmed 
    
def tokenize(text):
    lowers = text.lower()
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(lowers)
    return tokens 

def find_similar(doc,docs): 
    tfidf = CountVectorizer(tokenizer=tokenize,stop_words='english')
    tfidf = tfidf.fit_transform(doc)
    print(tfidf)
    
```

```python
#Demoncrat 
tokens_dem = tokenize(texts_dem)
filtered_words_dem = [word for word in tokens_dem if word not in stopwords.words('english')]
unigram_count_dem = Counter(filtered_words_dem)
for i in unigram_count_dem.most_common(100):
    print(i[0], i[1])
```

```python
#Republican 
tokens_rep = tokenize(texts_rep)
filtered_words_rep = [word for word in tokens_rep if word not in stopwords.words('english')]
unigram_count_rep = Counter(filtered_words_rep)
for i in unigram_count_rep.most_common(100):
    print(i[0], i[1])
```

```python
#Demoncrat 
bigram_dem = ngrams(filtered_words_dem,2)
bigram_count_dem = Counter(bigram_dem)
for i in bigram_count_dem.most_common(100):
    word_dem = ' '.join(i[0])
    print(word_dem, i[1])
```

```python
#Republican
bigram_rep = ngrams(filtered_words_rep,2)
bigram_count_rep = Counter(bigram_rep)
for i in bigram_count_rep.most_common(100):
    word_rep = ' '.join(i[0])
    print(word_rep, i[1])
```
